# Ephys Analysis

gets all imports and dataframes for ephys features and the labels. Merges them into one dataframe to match the indexes then creates a dataframe X, that contains the features used to predict labels.

```python
import pandas as pd
from get_features import get_dataframes, get_labels, get_morphology, get_all_features
from decisionTree import make_decisionTree, get_acc, plot_acc
```

```python
labels = get_labels()
epys_features = get_dataframes()
full_dataframe = pd.merge(epys_features, labels, left_index=True, right_index=True)
X = full_dataframe.drop(labels.columns, axis=1)


# plot accuracy using dendrite type feature
Y = full_dataframe["dendrite_type"]
labelName = "Dendrite Type"
classifiers = ["RandomForest", "DecisionTree"]
max_nodes = 30
featureType = "Electrophysiology"
plot_acc(X, Y, labelName, classifiers, featureType, max_nodes)

```

```python
# plot accuracy using structure area feature
Y = full_dataframe["structure_area_abbrev"]
labelName = "Brain Area"
classifiers = ["RandomForest", "DecisionTree"]
max_nodes = 30
featureType = "Electrophysiology"
plot_acc(X, Y, labelName, classifiers, featureType, max_nodes)
```

```python
full_dataframe["merge_all"] = full_dataframe["species"] + "_" + full_dataframe["dendrite_type"] + "_" + full_dataframe["structure_area_abbrev"] + "_" + full_dataframe["processed_structure_layer_name"]
X = full_dataframe.drop(labels.columns, axis=1)
X = X.drop(["merge_all"], axis = 1)
Y = full_dataframe["merge_all"]
labelName = "Merge All"
classifiers = ["RandomForest", "DecisionTree"]
max_nodes = 30
plot_acc(X, Y, labelName, classifiers,featureType, max_nodes)
```
