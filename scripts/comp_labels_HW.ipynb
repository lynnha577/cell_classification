```python
import pandas as pd
from get_features import get_dataframes, get_labels
from decisionTree import make_decisionTree, get_acc, plot_acc

labels = get_labels()
epys_features = get_dataframes()
full_dataframe = pd.merge(epys_features, labels, left_index=True, right_index=True)
X = full_dataframe.drop(labels.columns, axis=1)
X = X.drop(["id"], axis = 1)

```

```python
#plot accuracy using species feature
Y = full_dataframe["species"]
labelName = "species"
classifiers = ["RandomForest", "DecisionTree"]
max_nodes = 20
plot_acc(X, Y, labelName, classifiers, max_nodes)
```

```python
# plot accuracy using dendrite type feature
Y = full_dataframe["dendrite_type"]
labelName = "dendrite_type"
classifiers = ["RandomForest", "DecisionTree"]
max_nodes = 40
plot_acc(X, Y, labelName, classifiers, max_nodes)
# why does random forest train line stop at 30 nodes?
```

```python
# plot accuracy using structure area feature
Y = full_dataframe["structure_area_abbrev"]
labelName = "structure_area_abbrev"
classifiers = ["RandomForest", "DecisionTree"]
max_nodes = 40
plot_acc(X, Y, labelName, classifiers, max_nodes)
# why is test better than training for random forest classifier?
```

```python
# plot accuracy using layer name feature
Y = full_dataframe["processed_structure_layer_name"]
labelName = "processed_structure_layer_name"
classifiers = ["RandomForest", "DecisionTree"]
max_nodes = 20
plot_acc(X, Y, labelName, classifiers, max_nodes)
```

Out of the features, species, dendrite type, structure area abbrev, and layer name, the Random Forest model is the best at predicting the dentrite type and species features. It is not good at prediticting layer name and structure area abbrev features. I think that using the well predicted features will improve predictions but not the poorly predicted features.
