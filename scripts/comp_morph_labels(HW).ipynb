```python
from decisionTree import get_acc, make_decisionTree, decisionTreeAnalysis, plot_acc
from get_features import get_morphology, get_all_features, get_labels
import pandas as pd
from sklearn.ensemble import RandomForestClassifier


labels = get_labels()
morphology_features = get_morphology()
full_dataframe = pd.merge(morphology_features, labels, left_index=True, right_index=True)

X = full_dataframe.drop(labels.columns, axis=1)
classifiers = ["RandomForest", "DecisionTree"]
max_nodes = 24
```

```python
Y = full_dataframe["species"]
labelName = "species"
featureType = "morphology"
plot_acc(X, Y, labelName, classifiers, featureType, max_nodes)
```

```python
Y = full_dataframe["dendrite_type"]
labelName = "dendrite_type"
featureType = "morphology"
plot_acc(X, Y, labelName, classifiers, featureType, max_nodes)
```

```python
Y = full_dataframe["structure_area_abbrev"]
labelName = "structure_area_abbrev"
featureType = "morphology"
plot_acc(X, Y, labelName, classifiers, featureType, max_nodes)
```

```python
Y = full_dataframe["processed_structure_layer_name"]
labelName = "layer_number"
featureType = "morphology"
plot_acc(X, Y, labelName, classifiers, featureType, max_nodes)
```

the model is good at predicting species and dendrite type
the model is not good at predicting layer number.
It is the same as the ephys features because species and dendrite type is easy to predict because they have a higher chance accuracy.

```python
full_dataframe["dendrite_and_species"] = full_dataframe["dendrite_type"] + "_" + full_dataframe["species"]
X = full_dataframe.drop(labels.columns, axis=1)
X = X.drop(["dendrite_and_species"], axis = 1)
Y = full_dataframe["dendrite_and_species"]

labelName = "dendrite_and_species"
featureType = "morphology"
max_nodes = 40
plot_acc(X, Y, labelName, classifiers, featureType, max_nodes)
```

```python
full_dataframe["structure_area_with_dendrite"] = full_dataframe["structure_area_abbrev"] + "_" + full_dataframe["dendrite_type"]
X = full_dataframe.drop(labels.columns, axis=1)
X = X.drop(["layer_name_with_species", "dendrite_and_species", "layer_name_with_dendrite", "structure_area_with_dendrite"], axis = 1)
Y = full_dataframe["structure_area_with_dendrite"]

labelName = "structure_area_with_dendrite"
featureType = "morphology"
max_nodes = 40
plot_acc(X, Y, labelName, classifiers, featureType, max_nodes)
```
