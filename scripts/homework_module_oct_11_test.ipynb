```python
from get_features import get_labels
from plot_data import plot_morphology, plot_data
from allensdk.core.cell_types_cache import CellTypesCache

labels = get_labels()

def pick_specimen_id(species, dendrite):
    subset = labels[(labels['species'].str.lower() == species.lower()) & (labels['dendrite_type'] == dendrite)]
    if subset.empty:
        return None
    # index is the specimen id (the "id" field stored as index in get_labels)
    return int(subset.index[0])

sid_list = {
    "H_S": pick_specimen_id('Homo Sapiens', 'spiny'),
    "H_A": pick_specimen_id('Homo Sapiens', 'aspiny'),
    "M_S": pick_specimen_id('Mus musculus', 'spiny'),
    "M_A": pick_specimen_id('Mus musculus', 'aspiny')
}

SWEEP_NUMBER = 35   # choose sweep number to plot (change if not available)
TSTART = 0          # seconds (plot window start)
TEND = 30           # seconds (plot window end)

for label, s_id in sid_list.items():
    plot_morphology(s_id, )
    for sweep in range(10, 50):
        print(f"=== {label} - s_id ({s_id}) - {sweep} ===")
        try:
            plot_data(s_id, sweep, TSTART, TEND)
        except KeyError as e:
            print(f"NOT availabile: === {label} - s_id ({s_id}) - {sweep} ===")
            pass
```
